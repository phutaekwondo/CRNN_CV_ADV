```python
import matplotlib.pyplot as plt
import numpy as np
import os
import cv2

from pathlib import Path
```

```python
# Path to the data directory
data_dir = Path("./picked/")

# Get list of all the images
images = sorted(list(map(str, list(data_dir.glob("*.png")))))
labels = [img.split(os.path.sep)[-1].split(".png")[0] for img in images]
characters = set(char for label in labels for char in label)
characters = sorted(list(characters))

print("Number of images found: ", len(images))
print("Number of labels found: ", len(labels))
print("Number of unique characters: ", len(characters))
print("Characters present: ", characters)
```

```python
#show random 16 image

import random

# random 16 index from the dataset
samples_index = random.choices(range(len(images)), k=16)

plt.figure(figsize=(10, 10))

for i in range(16):
    plt.subplot(4, 4, i + 1)
    img = cv2.imread(images[samples_index[i]])
    plt.imshow(img)
    plt.title(labels[samples_index[i]])
    plt.axis("off")
```

```python
#amount of each character in the dataset
amount = {}
for label in labels:
    for char in label:
        if char in amount:
            amount[char] += 1
        else:
            amount[char] = 1

#sort amound by key
amount = {k: v for k, v in sorted(amount.items(), key=lambda item: item[1])}

#show bar chart
#size of the figure
plt.figure(figsize=(20, 10))
plt.bar(amount.keys(), amount.values())
plt.xlabel('Character')
plt.ylabel('Amount')
plt.show()

```
